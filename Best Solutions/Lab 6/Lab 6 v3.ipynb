```python
# Put your code here..
#
#
G = nx.karate_club_graph()
A = nx.to_numpy_matrix(G, dtype=int)
A = np.asarray(A)

def plotDist(A):
    
    f, ax = plt.subplots(2, 2, figsize=(10,10))
    ax[0, 0].imshow(A, cmap = 'Greys', interpolation = 'None')
    ax[0, 0].set_title('Adjacency Matrix')
    
    D = np.corrcoef(A)
    ax[1, 0].imshow(D, cmap = 'Greys', interpolation = 'None')
    ax[1, 0].set_title('Correlation coeff.')
    
    dVec = spt.distance.pdist(A, metric = 'euclidean')
    D = spt.distance.squareform(dVec)
    ax[0, 1].imshow(D, cmap = 'Greys', interpolation = 'None')
    ax[0, 1].set_title('Euclidean Dist.')
    
    dVec = spt.distance.pdist(A, metric = 'cosine')
    D = spt.distance.squareform(dVec)
    ax[1, 1].imshow(D, cmap = 'Greys', interpolation = 'None')
    ax[1, 1].set_title('Cosine Dist.')
    
plotDist(A)   
    
```

### Task 2 - Node Reordering

Without special preprocess procedures graph adjacency matrix can look very noisy and hide network's structure (just look at the matrices above). Offcourse usually you don't know the structure itself (communities, groups of closelly connected nodes, etc.) unless it is given, however there are some procedures of node reordering that provides a better view of the network's adjacency matrix.

*[Reverse Cuthill-McKee](http://en.wikipedia.org/wiki/Cuthill%E2%80%93McKee_algorithm)* finds permutation of the nodes that minimizes the **bandwidth** of the matrix, which is calculated as:
$$ \theta = \max_{a_{ij} > 0}|i-j|$$
Unformally, this algorithm puts some *mass* on the diagonal of adjacency matrix.

Run this reordering with *nx.utils.reverse_cuthill_mckee_ordering(G)* and compare with the results above

```python
# Put your code here
#
#
G = nx.karate_club_graph()

# run procedure
cm = nx.utils.reverse_cuthill_mckee_ordering(G)
# get permutation
l = [n for n in cm]
l

A = nx.to_numpy_matrix(G)
A = np.asarray(A)
# apply reordering
A = A[np.ix_(l,l)]

plotDist(A)

```

```python
kill_events = events[events['event'] == 'killed']
kill_events = pd.DataFrame(kill_events, index = None, columns=['characterID', 'event', 'withID'])
kill_events = kill_events.dropna()
```

```python
kill_events.head()
```

```python
G = nx.DiGraph()

for row, data in kill_events.iterrows():
    killer = data[2]
    killed = data[0]
    G.add_edge(killer, killed)
```

```python
ch_team = characters.set_index('characterID')['Team'].to_dict()
for k in ch_team.keys():
    if k not in G.nodes():
        del ch_team[k]
```

```python
nx.set_node_attributes(G, 'Team', ch_team)
```

### 2. Solution

```python
f = open("CollegeMsg.txt")
        
lines = []
degree_assortativity_coefficients = []
step = 1

for line in f:
    lines.append(line)
    
    if step % 250 == 0:
        G = nx.parse_edgelist(lines, data=(('timestamp', float), ))
        degree_assortativity_coefficients.append(nx.degree_assortativity_coefficient(G))
    step = step + 1
    
plt.plot(degree_assortativity_coefficients)
```

### Explanation:
+ Example can be found here: https://snap.stanford.edu/data/CollegeMsg.html
+ the graph above shows, that the degree assortativity decreases, as time advances.
+ this can be explained as follows:
    * in the beginning there was only a smaller number of people chatting with each other which had about the same degrees
    * as time progresses more people joined the network and the number of connected components started to increase
    * after a certain time the single connected components began to connect to each other, as users of different groups started sending each other messages
    * Later the degree assortativity is close to zero, meaning that nearly every node belongs to the big, single connected component which emerged.

### 3. Solution

```python
nodeList = G.nodes()

degree_assortativity_coefficients = [nx.degree_assortativity_coefficient(G)]
fitted_pl = powerlaw.Fit(G.degree().values(), discrete=True)
pl_xmin = [fitted_pl.xmin]
pl_alpha = [fitted_pl.alpha]
degree_tupels = G.degree().items()

sorted_degree = sorted(degree_tupels, key=lambda deg: deg[1], reverse=True)

for i in range(0,1500):
    nodeList.remove(sorted_degree[i][0])
    H = G.subgraph(nodeList)
    degree_assortativity_coefficients.append(nx.degree_assortativity_coefficient(H))
    fitted_pl = powerlaw.Fit(H.degree().values(), discrete=True)
    pl_xmin.append(fitted_pl.xmin)
    pl_alpha.append(fitted_pl.alpha)
```

```python
plt.plot(degree_assortativity_coefficients)
plt.title("Degree assortativity coefficient")
plt.show()
plt.plot(pl_xmin)
plt.title("xmin")
plt.show()
plt.plot(pl_alpha)
plt.title("alpha")
```

### Explanation (Example is the first one from http://biit.cs.ut.ee/graphweb/welcome.cgi?t=examples)
* As the above graph shows is the degree assortativity coefficient increasing, when the nodes with the max degree are deleted from the graph.
* Meanwhile, the power law values are constantly decreasing, as the the xmin and alpha graphs show.
* This stands in contradiction to the assumption of the task, but according to wikipedia are biological environments usually highly disassortative, confirming this example.
* When nodes with high degree of a biological network are deleted, the assortativity should increase, because these nodes most likely were connected to other nodes with low degrees.

#### Propose a generative model for disassortative growing scale-free networks.
* One way of generating such a network would be to check the degrees of all existing nodes in the network and inserting a new node to the node with the lowest degree. This way the network would be growing disassortative, because not the degree "rich" nodes are getting more edges, but the edge "poor" ones.
